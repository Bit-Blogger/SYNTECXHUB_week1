Average Marks

```python
subjects=['Math_Score',
          'Reading_Score',
          'Writing_Score']
df[subjects].mean()
```

Highest and Lowest Score

```python
df[subjects].max()
df[subjects].min()
```

```python
df['average_score'] = df[subjects].mean(axis=1)
df[['average_score']].head()
```

```python
df.sort_values(
    by='Average_Score',
    ascending=False
).head()
```

```python
df[['Study_Hours',
    'Average_Score']].corr()
```

```python
sns.scatterplot(

x='Study_Hours',
y='Average_Score',

data=df

)

plt.show()
```

```python
df[['Attendance',
'Average_Score']].corr()
```

```python
sns.scatterplot(

x='Attendance',
y='Average_Score',

data=df

)

plt.show()
```

```python
sns.boxplot(

x='Sleep_Hours',
y='Average_Score',

data=df

)

plt.show()
```

```python
df.groupby(

'Gender'

)['Average_Score'].mean()
```

```python
df.groupby(

'Class'

)['Average_Score'].mean()
```

```python
avg_subject = df[subjects].mean()


avg_subject.plot(kind='bar')


plt.title("Average Subject Scores")


plt.show()
```

```python
gender_count = df['Gender'].value_counts()



gender_count.plot(

kind='pie',

autopct='%1.1f%%'

)


plt.show()
```

```python
study_avg = df.groupby(

'Study_Hours'

)['Average_Score'].mean()


study_avg.plot()


plt.show()
```

```python
plt.hist(

df['Average_Score'],

bins=15

)

plt.show()
```

```python
corr = df.corr(numeric_only=True)



corr
```

```python
plt.figure(figsize=(8,6))


sns.heatmap(

corr,

annot=True,

cmap='coolwarm'

)


plt.show()
```

1. Students studying more hours tend to score better.
